```python
def crop_face(img, bb):
    """
    Crop image given bounding box (x,y, width, height)
    """
    x1, y1, width, height = bb
    x1,y1 = abs(x1), abs(y1)
    x2, y2 = x1 + width, y1 +height
    return img[y1:y2, x1:x2]
```

```python
def detect_faces(img, detector):
    results = detector.detect_faces(img)
    bb_lst =[]
    for i in results:
        bb_lst.append(i["box"])
    return bb_lst
```

```python
input_shape = m.input_shape[1:]
```

```python
input_shape
```

```python
for layer in m.layers[:-6]:
    layer.trainable = False
```

```python
def squared_differences(pair_of_tensors):
    x, y = pair_of_tensors
    return k.square(x - y)
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Input, Lambda, add

inputs1 = Input(input_shape)
inputs2 = Input(input_shape)
x1 = m(inputs1)
x2 = m(inputs2)
x1 = Model(inputs = inputs1, outputs = x1)
x2 = Model(inputs = inputs2, outputs = x2)


x = Lambda(squared_differences)([x1.output, x2.output])
output = Dense(1, activation = 'sigmoid')(x)
```

```python
model = Model(inputs =[x1.input,x2.input], outputs=output)
model.compile(
    optimizer = tf.keras.optimizers.Adam(0.01), 
    loss = 'binary_crossentropy', 
    metrics = ["accuracy"],
)
```

```python
def preprocess_image(image):
    
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [600, 600])
    image /= 255.0 

    return image
```

```python
def preprocess_image2(image):
    
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.resize(image, [600, 600])
#     image /= 255.0 

    return image
```

```python
def load_and_preprocess_image(path):
    image = tf.read_file(path)
    return preprocess_image(image)
```

```python
def load_and_preprocess_image2(path):
    image = tf.read_file(path)
    return preprocess_image2(image)
```

```python
def load_positive_match(label_index):
    label = index_to_label[label_index]
    dir_path = 'data/images/'+label
    img_path1 = dir_path+'/'+random.choice(os.listdir(dir_path))
    image = load_and_preprocess_image2(img_path1)
    return image
    
```

```python
import shutil

def load_random_match():
    dirs = os.listdir("data/images")
    if len(dirs) != 0:
        random_dir = random.choice(dirs)
        dir_files = os.listdir("data/images/"+random_dir)
        while len(dir_files) == 0:
            empty_dir = "data/images/"+random_dir
            #print('removing empty_dir: ', empty_dir)
            shutil.rmtree("data/images/"+random_dir)
            dirs = os.listdir("data/images")
            random_dir = random.choice(dirs)
            dir_files = os.listdir("data/images/"+random_dir)
        if len(dir_files) != 0:
            random_file = random.choice(dir_files)
            while '.jpg' not in random_file:
                #print('redo random.choice(dir_files) : ', random_dir)
                random_file = random.choice(dir_files)
            image = load_and_preprocess_image2("data/images/"+random_dir+'/'+random_file)
            return random_dir, image
        else:
            return '', None

    else:
        #print('dirs: ', dirs)
        return '', None
    
```
